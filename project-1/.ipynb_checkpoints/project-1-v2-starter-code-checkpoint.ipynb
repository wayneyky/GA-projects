## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def player_power(players_dict, pokedex_dict, player_id):
    """
    this function will go through each players pokemon(s) in the players dictionary and pick out the required stats and 
    append them into a separate list which i call att_def. so the total power of that player will just be the sum of 
    the numbers appended in att_def.
    
    players_dict will call the players dictionary
    pokedex_dict will call the pokedex dictionary
    player_id will call players["player_id"] in the players dictionary
    
    """
    att_def = []
    for pokemon in players[player_id]["player_pokemon"]:
        if pokemon in pokedex:
            # i have no idea whih stats are base stats so i just add the special att and def too, I nvred played pokemon before hah.
            att_def.append(pokedex[pokemon]["attack"])
            att_def.append(pokedex[pokemon]["defense"])
            att_def.append(pokedex[pokemon]["special_attack"])
            att_def.append(pokedex[pokemon]["special_defense"])
    return sum(att_def)
print(player_power(players, pokedex, 1), player_power(players, pokedex, 2))
````

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

````python
clean1 = raw_pd.splitlines()
````

````python
# trying to remove the quote marks and split the values on comma
pokedex_units = []
for header in clean1:
    pokemon_unit = header.split(",")
    pokemon_unit = [(poke.replace('"', '')) for poke in pokemon_unit]
    pokedex_units.append(pokemon_unit)

# filter the floats from stuff that are not floats using another dummy list 
cleanraw = []
for pokemon in pokedex_units:
    floats = []
    for stat in pokemon: 
        try:
            floats.append(float(stat))
        except:
            floats.append(stat)
    cleanraw.append(floats)
print(cleanraw)
````

## 9. Write a function to generate the full pokedex

---

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. The `PokedexNumber` should be used as the `pokemon_id` key values for the dictionary of pokemon.

Your function should:

1. Take the parsed pokedex information you created above as an argument.
2. Return a dictionary in the same format as your original pokedex you created before containing the information from the parsed full pokedex file.

To test the function, print out the pokemon with id = 100.

````python
# this function will create a pokedex and give u the pokemon when u enter the raw data and pokemon id number

def create_pokedex(pokemon_dataset, pokemon_id):
    dataset = pokemon_dataset[1:]
    header = pokemon_dataset[0]
    pokemon_id = float(pokemon_id)
    pokedex_data = []
    for number in range(800):
        data = dataset[number]
        #print(data)
        pokedex_unit = dict(zip(header[1:], data[1:]))
        pokedex_data.append(pokedex_unit)
    #print(pokedex_data)
    pokedex_v0 = {pokemon_number[0]:pokedex_data[dataset.index(pokemon_number)] for pokemon_number in dataset}
    for pokemon in pokedex_v0:
        if pokemon_id == pokemon:
            return pokedex_v0[pokemon_id]
print(create_pokedex(cleanraw, 100))
````

````python
def filtered_pokedex(pokedex_data, filter):
    filtered_pokemon = []
    for pokemon in pokedex_data:
        if pokedex_data[pokemon]['Attack'] >= filter['Attack'] and pokedex_data[pokemon]['Defense'] >= filter['Defense'] and pokedex_data[pokemon]['Type'] == filter['Type']:
            filtered_pokemon.append(pokedex_data[pokemon])
    return filtered_pokemon
filtered_pokedex(pokedex_v1, {'Attack':25, 'Defense':30, 'Type':'Electric'})
````

````python
def filtered_pokedex(pokedex_data, filter_options):
    pokedex_data = [pokedex_data[pokemon] for pokemon in pokedex_data]
    #filtering_process
    filter_type = [attribute for attribute in filter_options if attribute == 'Type']
    filter_name = [attribute for attribute in filter_options if attribute == 'Name']
    filter_therest = [attribute for attribute in filter_options if attribute != 'Type' and attribute != 'Name']
    if len(filter_type) == 0 and len(filter_name) == 0:
        for therest in filter_therest:
            results = [data_therest for data_therest in pokedex_data if data_therest[therest] >= filter_options[therest]]
            return results
    elif len(filter_type) == 0:
        for pokemon_name in filter_name:
            results_name = [data_name for data_name in pokedex_data if (data_name[pokemon_name] in filter_options[pokemon_name])]
        for therest in filter_therest:
            results = [data_therest for data_therest in results_name if data_therest[therest] >= filter_options[therest]]
            return results
    elif len(filter_name) == 0:
        for pokemon_type in filter_type:
            results_type = [data_type for data_type in pokedex_data if data_type[pokemon_type] == filter_options[pokemon_type]]
        for therest in filter_therest:
            results = [data_therest for data_therest in results_type if data_therest[therest] >= filter_options[therest]]
            return results
    else:
        for pokemon_name in filter_name:
            results_name = [data_name for data_name in pokedex_data if (data_name[pokemon_name]) in filter_options[pokemon_name]]
        for pokemon_type in filter_type:
            results_type = [data_type for data_type in results_name if data_type[pokemon_type] == filter_options[pokemon_type]]
        for therest in filter_therest:
            results = [data_therest for data_therest in results_type if all(data_therest[therest] >= filter_options[therest])]
            return results
print(filtered_pokedex(pokedex_v1, {'Attack':95,'Defense':95,'Type':'Poison','SpecialAttack':80, 'SpecialDefense':70}))
````
